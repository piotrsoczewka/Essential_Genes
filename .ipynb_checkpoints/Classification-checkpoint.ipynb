```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stat
```

```python
# importing data
all_genes_df = pd.read_csv('all_genes_df.csv')
all_genes_df.head()
```

```python
# creating lists of essential and non essential genes form txt files
essential_genes = []
with open('essential_genes.txt') as file:
    for line in file:
        line_index_0 = line.split('\t')[0]
        if line_index_0[0] == 'Y':
            essential_genes.append(line_index_0)

            
non_essential_genes = []
with open('non_essential_genes.txt', encoding="utf8") as file:
    for line in file:
        line_index_0 = line.split('\t')[0]
        if line_index_0[0] == 'Y':
            non_essential_genes.append(line_index_0)
```
